**o	Create any other metric apart from (Level5 visitors/Level1 visitors) and perform the same task as above. Compare if the states are same in both the questions, if they are different try to create a hypothesis about the reason behind it.**

```python
data_frame1["Total Visitors"] = data_frame1["Lv1_Visitors"] + data_frame1["Lv2_Visitors"] + data_frame1["Lv3_Visitors"] + data_frame1["Lv4_Visitors"] + data_frame1["Lv5_Visitors"]

# calculate new metric
data_frame1["New Metric"] = data_frame1["Lv5_Visitors"] / data_frame1["Total Visitors"]

# top 3 states based on new metric for each year and segment
for year in data_frame1["Year"].unique():
    for segment in data_frame1["Segment"].unique():
        temp_data_frame = data_frame1[(data_frame1["Year"]==year) & (data_frame1["Segment"]==segment)]
        temp_data_frame = temp_data_frame.groupby(["Region"]).agg({"New Metric": "mean"}).reset_index()
        temp_data_frame = temp_data_frame.sort_values("New Metric", ascending=False).head(3)
        print(f"Top 3 states for {segment} in {year}:")
        print(temp_data_frame)
        print("\n")
```

The variation in the top-performing regions between the new metric and the Level 5 visitors/Level 1 visitors metric can be attributed to the differences in their calculations. The new metric takes into account the total number of visitors in each segment and region, providing a more comprehensive measure of visitor engagement and performance.

Unlike the Level 5 visitors/Level 1 visitors metric, which focuses solely on the ratio of Level 5 to Level 1 visitors, the new metric considers the overall popularity and attractiveness of a region by incorporating the total visitor counts. As a result, regions with high overall visitor numbers, even if they have a relatively lower proportion of Level 5 visitors, can still rank higher in the new metric. Conversely, regions with a high ratio of Level 5 visitors to Level 1 visitors may not rank as high if their total visitor counts are lower.

```python
import statsmodels.api as sm
data_frame3['Date'] = pd.to_datetime(data_frame3['Year'].astype(str) + '-' + data_frame3['Month'].astype(str))
data_frame3 = data_frame3.set_index('Date')
model = sm.tsa.ARIMA(data_frame3['Lv5_Visitors'], order=(1,0,50))
results = model.fit()
forecast_values = results.forecast(steps=6)
forecast_index = pd.date_range(data_frame3.index[-1], periods=6, freq='M')

# Set the forecast index to dates
forecast_values.index = forecast_index
```
